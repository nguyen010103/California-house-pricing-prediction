```python
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing(as_frame=True)

df = california_housing.frame
df.head()
```

```python
#understand the data
df.info
```

```python
df.dtypes
```

```python
#check if they have missing value or not
df.isnull().sum()
```

```python
#rename column for better 
df.columns = [col.lower() for col in df.columns]
df.rename(columns = {
    'medinc':'median_income',
    'houseage': 'housing_age',
    'averooms': 'avg_rooms',
    'avebedrms': 'avg_bedrooms',
    'population': 'population',
    'aveoccup': 'avg_occupants',
    'latitude': 'latitude',
    'longitude': 'longitude',
    'medhouseval': 'median_house_value'
}, inplace=True)
```

Column Name	Description

median_income - Median income of households in the block group (scaled, e.g., 3.5 = $35,000)

housing_age - Median age of houses in the block group (in years)

avg_rooms - Average number of rooms per household

avg_bedrooms - Average number of bedrooms per household

population - Total population in the block group

avg_occupants - Average number of occupants per household

latitude - Latitude of the block group’s geographic center

longitude - Longitude of the block group’s geographic center

median_house_value - Median house value in the block group (in $100,000 units — maxed at $500,000)
